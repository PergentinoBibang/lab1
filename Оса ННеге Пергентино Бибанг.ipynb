## Determine the class with the longest working hours

```python
df.head()
```

```python
ax=df.query('Workclass != "Self-emp-not-inc"')\
.groupby('Workclass')['Hours_per_week']\
.agg(['mean','count'])\
.query('count >= 100')\
.sort_values('mean') ['mean']\
.plot(kind='barh', figsize=(12,5), title='Average Weekly working Hours')
ax.set_xlabel('Average Weekly working hours')
plt.show()
```

# Conclusion

#### After getting the graph its logical and normal to conclude that Self-emp-incs have the longest average working hours in a week.
